```python
import numpy as np
import matplotlib.pyplot as plt
import iisignature as iis
import networkx as nx
```

```python
def lead_matrix_1(mv_time_series):
    N=mv_time_series.shape[1]
    sig=iis.sig(mv_time_series,2)
    S=sig[N:].reshape(N,N)
    L=(S-S.T)/2
    return L
    
```

```python
def lead_tensor_1(mv_time_series):
    N=mv_time_series.shape[1]
    sig=iis.sig(mv_time_series,3)
    l=N+N*N
    S=sig[l:].reshape(N,N,N)
    T=(S-S.transpose((0,2,1))+S.transpose((2,0,1))-S.transpose((2,1,0))+S.transpose((1,2,0))-S.transpose((1,0,2)))/6
    return T
```

```python
# FIRST EXPERIMENT



t=np.linspace(0,49,100)
x=8*np.sin(t)
y=8*np.sin(t-0.2)
z=8*np.sin(t-0.4)
w=8*np.sin(t-0.6)

a=6*np.sin(0.5*t+2.4)
b=6*np.sin(0.5*t+2.2)
c=6*np.sin(0.5*t+2)


plt.plot(t,x)
plt.plot(t,y)
plt.plot(t,z)
plt.plot(t,w)
plt.plot(t,a)
plt.plot(t,b)
plt.plot(t,c)
plt.legend(["x", "y", "z", "w", "a", "b", "c"])
plt.show()


time_series=np.zeros((100,7))
time_series[:,0]=x
time_series[:,1]=y
time_series[:,2]=z
time_series[:,3]=w
time_series[:,4]=a
time_series[:,5]=b
time_series[:,6]=c
```

```python
L=lead_matrix_1(time_series)
plt.imshow(L, alpha=0.9, cmap="Blues")
plt.colorbar()
```

```python
T=lead_tensor_1(time_series)

fig, axs = plt.subplots(2, 4, figsize=(20,17), sharex=True, sharey=True)

for i in range(7):
    
    ax = axs.flatten()
    
    pos=ax[i].matshow(T[i,:,:], alpha=0.9, cmap="Blues")
    pos
    fig.colorbar(pos,ax=ax[i])
```
